```python
from sklearn.metrics import classification_report, confusion_matrix
import os
import glob
from PIL import Image
```

```python
# Path to the validation dataset
val_dir = '/kaggle/working/YOLO/val'

# Function to get the ground truth labels
def get_ground_truth_labels(val_dir):
    true_labels = []
    true_label_names = []

    for class_folder in os.listdir(val_dir):
        folder_path = os.path.join(val_dir, class_folder)
        if os.path.isdir(folder_path):
            # All images in this folder have the same true label
            for image_name in os.listdir(folder_path):
                if image_name.endswith(('.png', '.jpg', '.jpeg')):  # Check if it's an image file
                    true_labels.append(class_folder)
                    true_label_names.append(class_folder)  # Storing the label name

    return true_labels, true_label_names

# Get the ground truth labels
true_labels, true_label_names = get_ground_truth_labels(val_dir)

# Print the ground truth labels
print("Ground truth labels:")
print(true_labels)
```

```python
model=YOLO('/kaggle/input/model-yolo/kaggle/working/runs/classify/train/weights/best.pt')
```

```python
from sklearn.metrics import classification_report
from pathlib import Path

# Assuming you have a base path where all class folders are located
base_path = Path('/kaggle/working/YOLO/val')

# List of all class folder names
class_folders = [folder.name for folder in base_path.glob('*') if folder.is_dir()]

# Initialize lists to store true and predicted labels
true_labels = []
predicted_labels = []

# Loop through each class folder and make predictions on each image
for class_folder in class_folders:
    print('class folder => :', class_folder)

    folder_path = base_path / class_folder
    print('folder path =>', folder_path)
    
    # Make predictions on the current folder
    results = model(source=str(folder_path), conf=0.25)

    # Process each result
    for result in results:
        # Get the ground truth from the folder name
        true_labels.append(class_folder)

        # Get the predicted class name using the top1 index
        predicted_class_name = result.names[result.probs.top1]
        print('predicted_class_name =>', predicted_class_name)
        predicted_labels.append(predicted_class_name)

# Now that we have all the true labels and predicted labels, we can calculate the classification report
report = classification_report(true_labels, predicted_labels, target_names=class_folders,digits=4)
print(report)
```
